```python
# Get the HTML text from the wiki page
CanM_url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
CanM_text = requests.get(CanM_url).text
html_text
```

Extract the table data from the HTML text using BeautifulSoup

```python
soup = BeautifulSoup(CanM_text)
# In the html text, 'tr' defines the row of cells in a table.  Use 'tr' to get the number of rows in the table
table = soup.find('table')                                         #, attrs={'class':'wikitable sortable'})
trs = table.find_all('tr')
print("Number of rows including columns header: ",len(trs))

# Create an empty list of rows. Then extract text from all the table cells to add to rows.
# Use 'td' tag to find data cells.
rows = list()
for tr in trs:
    td = tr.find_all('td')   # Use 'td' tag to find data cells
    row = [ele.text.strip() for ele in td]
    if row:
        rows.append(row)

print("Number of rows with data: ", len(rows))
print(rows)  # To make sure necessary information has made it to the rows.
```

Once satisfies that rows look OK, create a pandas dataframe with columns 'PostalCode', 'Borough', and 'Neighborhood'.

```python
df = pd.DataFrame(rows,
                  columns=['PostalCode', 'Borough', 'Neighborhood'])
print("Dataframe shape: ", df.shape)
df.head(10)
```

Remove rows with column 'Not assigned' for column 'Borough'.

```python
df = df[df.Borough != 'Not assigned']
df.reset_index(inplace=True, drop=True)  #Reset the index in the table after dropping rows from the previous step
df.shape  
```

```python
df.head(10)
```

Assignment states that if a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough. To do that, substitute 'Neighborhood' columns with value as 'Not assigned' with the value of its 'Borough'.

```python
df['Neighborhood'] = df.apply(
    lambda row: 
    row['Borough'] if row['Neighborhood'] == 'Not assigned' 
    else row['Neighborhood'],
    axis=1)
#Print all rows in df to make sure that the above problem has been resolved
# To be able to see all rows, use the following option and set max_rows to none.

pd.set_option('display.max_rows', None)
df.head(len(rows))

```

#### Explore first neighborhood in Toronto and get top 50 avenues 

```python
neighborhood_name = df.loc[0, 'Neighborhood']
neighborhood_latitude = df.loc[0, 'Latitude']
neighborhood_longitude = df.loc[0, 'Longitude']

print("Latitude and longitude of neighborhood '{}' are [{}, {}]".format(
    neighborhood_name, neighborhood_latitude, neighborhood_longitude))

# Construct the URL
limit = 100
radius = 5000
explore_url_prefix = 'https://api.foursquare.com/v2/venues/explore'
url = '{}?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    explore_url_prefix, CLIENT_ID, CLIENT_SECRET, VERSION, 
    neighborhood_latitude, neighborhood_longitude, radius, limit)

# Get the venues.
results = requests.get(url).json()
results
```

```python
# Explore the venues
venues = results['response']['groups'][0]['items']
venues
```

```python
# Normalize the JSON response
neighborhood_venues = pd.json_normalize(venues)
neighborhood_venues
```

#### Keep only the columns needed for remaining exercise and simplify the names of the columns

```python
keep_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
neighborhood_venues = neighborhood_venues.loc[:, keep_columns]

neighborhood_venues.columns = [column.split(".")[-1] for column in neighborhood_venues.columns]
neighborhood_venues
```

```python
# Extract the category names from rows and replace the contents of the category column with just the category names

def get_catname(row):
    cat_list = row['categories']
    if cat_list:
        return cat_list[0]['name']
    return None

neighborhood_venues['categories'] = neighborhood_venues.apply(get_catname, axis=1)
neighborhood_venues
```

#### Repeat above steps for all neighborhoods

```python
venues_list = list()

for name, lat, lng in zip(df['Neighborhood'], df['Latitude'], df['Longitude']):
    print("Collecting venues for neighborhood:", name)
    
    # Create API request URL
    limit = 100
    radius = 5000
    explore_url_prefix = 'https://api.foursquare.com/v2/venues/explore'
    url = '{}?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
        explore_url_prefix, CLIENT_ID, CLIENT_SECRET, VERSION, 
        lat, lng, radius, limit)
    
    # Make the request
    neighborhood_venues = requests.get(url).json()["response"]['groups'][0]['items']
    
    # Add relevant info to venues_list
    venues_list.extend([(
        name, lat, lng,
        v['venue']['name'], 
        v['venue']['location']['lat'], 
        v['venue']['location']['lng'],  
        v['venue']['categories'][0]['name']) for v in neighborhood_venues])

print(len(venues_list))   
```

```python
#convert venues list to a datafreame

dfTorVenues = pd.DataFrame(venues_list)
dfTorVenues.columns = [
    'Neighborhood', 'Neighborhood Latitude', 'Neighborhood Longitude', 
    'Venue', 'Venue Latitude', 'Venue Longitude', 'Venue Category']
print(dfTorVenues.shape)
dfTorVenues.head(20)
```

```python
# Use one hot encoding.  If a neighborhood has a venue category, then Venue Category = 1, else = 0.
Toronehot = pd.get_dummies(dfTorVenues[['Venue Category']], prefix="", prefix_sep="")
print(Toronehot.head(10))

# Add neighborhood column back to dataframe
Toronehot.insert(0, 'NeighborhoodName', dfTorVenues['Neighborhood']) 
print(Toronehot.head(10))

```

```python
# Group neighborhoods and calculate mean frequency of each category.

Torgrouped = Toronehot.groupby('NeighborhoodName').mean().reset_index()
Torgrouped

```

#### Create a dataframe with 5 most common categories for each neighborhood

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 5

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = Torgrouped['NeighborhoodName']

for ind in np.arange(Torgrouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(
        Torgrouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

#### Cluster Neighborhoods.  Run k means and create 4 clusters

```python
# Set number of clusters
k = 4

# Drop the neighborhood name column so that each column contains only the feature set.
Torgrouped_clustering = Torgrouped.drop('NeighborhoodName', 1)

# Run k-means clustering
kmeans = KMeans(n_clusters=k, random_state=0).fit(Torgrouped_clustering)

# Check cluster labels generated for each row in the dataframe
kmeans.labels_[0:100] 
```

```python
# Add clustering lables


neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

dfmerged = df

# Merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
dfmerged = dfmerged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

print(dfmerged.shape)
dfmerged.head(10)
```

#### Map clusters

```python
map_toronto = folium.Map(
    location=[location.latitude, location.longitude],
    zoom_start=11)

# Set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# Add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(
        dfmerged['Latitude'], dfmerged['Longitude'],
        dfmerged['Neighborhood'], dfmerged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_toronto)
       
map_toronto
```
